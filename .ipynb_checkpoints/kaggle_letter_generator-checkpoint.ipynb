```python
# Function for processing an image
def image_to_tensor(img_path, folder_path):
    img = keras_image.load_img(folder_path + img_path, target_size=(32, 32))
    x = keras_image.img_to_array(img)
    return np.expand_dims(x, axis=0)
# Function for creating the data tensor
def data_to_tensor(img_paths, folder_path):
    list_of_tensors = [image_to_tensor(img_path, folder_path) for img_path in tqdm(img_paths)]
    return np.vstack(list_of_tensors)
ImageFile.LOAD_TRUNCATED_IMAGES = True
```

```python
# Load the data
data = pd.read_csv("/Users/olgabelitskaya/deep_learning_projects/DL_PP2/data2/letters2.csv")

files = data['file']
letters = data['letter']
tensors = data_to_tensor(files, "/Users/olgabelitskaya/deep_learning_projects/DL_PP2/data2/");
```

```python
# Create random images
def latent_samples(n_samples, sample_size):
    return np.random.normal(loc=0, scale=1, size=(n_samples, sample_size))
latent_sample_1024 = latent_samples(1, 1024)
latent_sample_128 = latent_samples(1, 128)
```

```python
def preprocess(x):    
    x = (x - 0.25) * 2
    x = np.clip(x, -1, 1)
    return x
def deprocess(x):
    x = (x / 2 + 0.5) * 255
    x = np.clip(x, 0, 255)
    x = np.uint8(x)
    x = x.reshape(32, 32)
    return x
```

```python
X_train_real = preprocess(X_train_real)
X_test_real  = preprocess(X_test_real)
```

```python
def display_images(generated_images):
    n_images = len(generated_images)
    rows = 4
    cols = n_images//rows
    
    plt.figure(figsize=(cols, rows))
    for i in range(n_images):
        img = deprocess(generated_images[i])
        plt.subplot(rows, cols, i+1)
        plt.imshow(img, cmap=cm.bone)
        plt.xticks([])
        plt.yticks([])
    plt.tight_layout()
    plt.show();
```

## Keras GAN
[`"Using GAN for Generating Hand-written Digit Images" by Naoki Shibuya`](https://github.com/naokishibuya/deep-learning/blob/master/python/gan_mnist.ipynb)

```python
discriminator = Sequential([Dense(128, input_shape=(1024,)),
                            LeakyReLU(alpha=0.01),
                            Dense(1),
                            Activation('sigmoid')], 
                           name='discriminator')

discriminator.summary()
```

```python
generator = Sequential([Dense(1024, input_shape=(128,)),
                        LeakyReLU(alpha=0.01),
                        Dense(1024),
                        Activation('tanh')], 
                       name='generator')

generator.summary()
```

```python
plt.figure(figsize=(3,3))
generated_latent_sample = generator.predict(latent_sample_128)
plt.title(discriminator.predict(generated_latent_sample))
plt.imshow(generated_latent_sample.reshape(32, 32), cmap=cm.bone);
```

```python
gan = Sequential([generator,discriminator])
gan.summary()
```

```python
def trainable(model, trainable):
    for layer in model.layers:
        layer.trainable = trainable
```

```python
def simple_GAN(sample_size, 
               g_hidden_size, 
               d_hidden_size, 
               leaky_alpha, 
               g_learning_rate,
               d_learning_rate):
    
    backend.clear_session()
    
    generator = Sequential([Dense(g_hidden_size, input_shape=(sample_size,)),
                            LeakyReLU(alpha=leaky_alpha),
                            Dense(1024), Activation('tanh')], 
                           name='generator')    

    discriminator = Sequential([Dense(d_hidden_size, input_shape=(1024,)),
                                LeakyReLU(alpha=leaky_alpha),
                                Dense(1),
                                Activation('sigmoid')], 
                               name='discriminator')    
    
    gan = Sequential([generator, discriminator])
    
    discriminator.compile(optimizer=Adam(lr=d_learning_rate), loss='binary_crossentropy')
    gan.compile(optimizer=Adam(lr=g_learning_rate), loss='binary_crossentropy')
    
    return gan, generator, discriminator
```

```python
def real_fake_labels(size):
    return np.ones([size, 1]), np.zeros([size, 1])
```

```python
y_real5, y_fake5 = real_fake_labels(5)
print('Real\n', y_real5)
print('Fake\n', y_fake5)
```

```python
sample_size     = 128     
g_hidden_size   = 1024     # generator
d_hidden_size   = 128     # discriminator
leaky_alpha     = 0.01
g_learning_rate = 0.0002  # generator
d_learning_rate = 0.002   # discriminator
epochs          = 1000
batch_size      = 64      
valid_size      = 16     
smooth          = 0.1
```

```python
y_train_real, y_train_fake = real_fake_labels(batch_size)
y_valid_real,  y_valid_fake  = real_fake_labels(valid_size)

gan, generator, discriminator = simple_GAN(sample_size, 
                                           g_hidden_size, 
                                           d_hidden_size, 
                                           leaky_alpha, 
                                           g_learning_rate,
                                           d_learning_rate)
```

```python
losses = []
for e in range(epochs):
    for i in range(len(X_train_real)//batch_size):
        # real MNIST digit images
        X_batch_real = X_train_real[i*batch_size:(i+1)*batch_size]
        
        # latent samples and the generated digit images
        batch_latent_samples = latent_samples(batch_size, sample_size)
        X_batch_fake = generator.predict_on_batch(batch_latent_samples)
        
        # train the discriminator to detect real and fake images
        trainable(discriminator, True)
        discriminator.train_on_batch(X_batch_real, y_train_real * (1 - smooth))
        discriminator.train_on_batch(X_batch_fake, y_train_fake)

        # train the generator via GAN
        trainable(discriminator, False)
        gan.train_on_batch(batch_latent_samples, y_train_real)
    
    # evaluate
    X_valid_real = X_test_real[np.random.choice(len(X_test_real), valid_size, replace=False)]
    
    valid_latent_samples = latent_samples(valid_size, sample_size)
    X_valid_fake = generator.predict_on_batch(valid_latent_samples)

    d_loss  = discriminator.test_on_batch(X_valid_real, y_valid_real)
    d_loss += discriminator.test_on_batch(X_valid_fake, y_valid_fake)
    g_loss  = gan.test_on_batch(valid_latent_samples, y_valid_real) 
    
    losses.append((d_loss, g_loss))
    
    if (e+1)%50==0:
        print("Epoch: {:>3}/{} | Discriminator Loss: {:>6.4f} | Generator Loss: {:>6.4f} | DL > GL: {}"\
              .format(e+1, epochs, d_loss, g_loss, d_loss>g_loss))
    
generator.save('letter_generator.h5')
```

```python
def display_loss(losses, n):
    indices = [i*n for i in range(len(losses)//n)]
    n_losses = np.array(losses)[indices, :]    
    plt.figure(figsize=(18, 6))
    plt.plot(n_losses.T[0], '-o', c='#37c9e1', lw=1, label='Discriminator')
    plt.plot(n_losses.T[1], 
             '-o', c='#39d4be', lw=1, label='Generator')
    plt.title("Training Loss Functions")
    plt.legend();
```

```python
display_loss(losses, 10)
```

## Keras DCGAN
[`"Deep Convolutional GAN (DCGAN) with MNIST" by Naoki Shibuya`](https://github.com/naokishibuya/deep-learning/blob/master/python/dcgan_mnist.ipynb) 

```python
X_train_real = X_train_real.reshape(-1, 32, 32, 1)
X_test_real = X_test_real.reshape(-1, 32, 32, 1)
```

```python
def dcgan_generator(input_size, leaky_alpha):
    # Input: 32,32,1
    return Sequential([
        Dense(1024, input_shape=(input_size,)),
        Reshape(target_shape=(4, 4, 64)),
        BatchNormalization(),
        LeakyReLU(alpha=leaky_alpha),
        # Output: 4,4,64
        
        Conv2DTranspose(128, kernel_size=5, strides=2, padding='same'), 
        BatchNormalization(),
        LeakyReLU(alpha=leaky_alpha),        
        # Output: 8,8,128
        
        Conv2DTranspose(256, kernel_size=5, strides=2, padding='same'), 
        BatchNormalization(),
        LeakyReLU(alpha=leaky_alpha),
        # Output: 16,16,256

        Conv2DTranspose(1, kernel_size=5, strides=2, padding='same'),
        Activation('tanh')
        # Output: 32,32,1
    ])
```

```python
dcgan_generator_example = dcgan_generator(1024, 0.01)
dcgan_generator_example.summary()
```

```python
def dcgan_discriminator(leaky_alpha):
    # Input: 32,32,1
    return Sequential([        
        Conv2D(256, kernel_size=5, strides=2, padding='same', input_shape=(32,32,1)),
        LeakyReLU(alpha=leaky_alpha),
        # Output: 16,16,256

        Conv2D(128, kernel_size=5, strides=2, padding='same'),   
        BatchNormalization(),
        LeakyReLU(alpha=leaky_alpha),
        # Output: 8,8,128
        
        Conv2D(64, kernel_size=5, strides=2, padding='same'),   
        BatchNormalization(),
        LeakyReLU(alpha=leaky_alpha),
        # Output: 4,4,64
        
        Flatten(),
        Dense(1024),
        BatchNormalization(),
        LeakyReLU(alpha=leaky_alpha),
        # Output: 1024
        
        Dense(1),
        Activation('sigmoid')
        # Output: 1
    ])
```

```python
dcgan_discriminator_example = dcgan_discriminator(0.01)
dcgan_discriminator_example.summary()
```

```python
def dcgan(sample_size, 
          g_learning_rate, g_beta_1,
          d_learning_rate, d_beta_1,
          leaky_alpha):
    # clear first
    backend.clear_session()
    
    # generator
    generator = dcgan_generator(sample_size, leaky_alpha)

    # discriminator
    discriminator = dcgan_discriminator(leaky_alpha)
    discriminator.compile(optimizer=Adam(lr=d_learning_rate, beta_1=d_beta_1), loss='binary_crossentropy')
    
    # GAN
    gan = Sequential([generator, discriminator])
    gan.compile(optimizer=Adam(lr=g_learning_rate, beta_1=g_beta_1), loss='binary_crossentropy')
    
    return gan, generator, discriminator
```

```python
def dcgan_train(g_learning_rate, g_beta_1,
                d_learning_rate, d_beta_1,
                leaky_alpha, 
                smooth=0.1,
                sample_size=128,
                epochs=30,
                batch_size=64,
                valid_size=16,
                show_details=True):
    
    # labels for the batch size and the test size
    y_train_real, y_train_fake = real_fake_labels(batch_size)
    y_valid_real,  y_valid_fake  = real_fake_labels(valid_size)

    # create a GAN, a generator and a discriminator
    gan, generator, discriminator = dcgan(sample_size, 
                                          g_learning_rate, g_beta_1,
                                          d_learning_rate, d_beta_1,
                                          leaky_alpha)

    losses = []
    for e in range(epochs):
        for i in range(len(X_train_real)//batch_size):
            # real MNIST digit images
            X_batch_real = X_train_real[i*batch_size:(i+1)*batch_size]

            # latent samples and the generated digit images
            batch_latent_samples = latent_samples(batch_size, sample_size)
            X_batch_fake = generator.predict_on_batch(batch_latent_samples)

            # train the discriminator to detect real and fake images
            trainable(discriminator, True)
            discriminator.train_on_batch(X_batch_real, y_train_real * (1 - smooth))
            discriminator.train_on_batch(X_batch_fake, y_train_fake)

            # train the generator via GAN
            trainable(discriminator, False)
            gan.train_on_batch(batch_latent_samples, y_train_real)

        # evaluate
        X_valid_real = X_test_real[np.random.choice(len(X_test_real), valid_size, replace=False)]

        valid_latent_samples = latent_samples(valid_size, sample_size)
        X_valid_fake = generator.predict_on_batch(valid_latent_samples)

        d_loss  = discriminator.test_on_batch(X_valid_real, y_valid_real)
        d_loss += discriminator.test_on_batch(X_valid_fake, y_valid_fake)
        g_loss  = gan.test_on_batch(valid_latent_samples, y_valid_real) 

        losses.append((d_loss, g_loss))

        print("Epoch:{:>3}/{} | Discriminator Loss:{:>7.4f} | Generator Loss:{:>7.4f} | DL > GL: {}"\
              .format(e+1, epochs, d_loss, g_loss, d_loss>g_loss))
        
        if show_details and (e+1)%3==0: display_images(X_valid_fake)
            
    if show_details: display_loss(losses, 1)
        
    return generator
```

```python
dcgan_train(g_learning_rate=0.0002, g_beta_1=0.8, 
            d_learning_rate=0.001, d_beta_1=0.8, 
            leaky_alpha=0.01);
```
